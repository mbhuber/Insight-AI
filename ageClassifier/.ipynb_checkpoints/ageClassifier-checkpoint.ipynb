```python
import os, shutil
import numpy as np
import time
import matplotlib.pyplot as plt

from keras import applications
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Lambda
from keras.layers import Activation, Dropout, Flatten, Dense, BatchNormalization
from keras.preprocessing.image import ImageDataGenerator
import keras.backend as K

from keras.callbacks import TensorBoard

K.set_image_dim_ordering('th')

# sklearn
# from sklearn.metrics import accuracy_score, confusion_matrix

%matplotlib inline
```

### Data

```python
def createImageGenerator(imageDir,batchSize,targetSize):
    return ImageDataGenerator().flow_from_directory(
        imageDir,  # this is the target director
        target_size=targetSize,  # all images will be resized to 150x150
        batch_size=batchSize,
        class_mode='categorical')  # since we use binary_crossentropy loss, we need binary labels

def preprocessImageBatch(imaArr):
    imaArr = imaArr.transpose((0,3,1,2))
    imaArr = (imaArr.astype(np.float32) - 127.5) / 127.5
    return imaArr
```

```python
pipePath = '../../data/imdb-wiki_crop_clean_align128_kerasPipe/'
trainDir = pipePath + "train/"
testDir  = pipePath + "test/"
```

```python
batch_size = 32

train_generator = createImageGenerator(trainDir,batch_size,(imaDim,imaDim))
test_generator  = createImageGenerator(testDir,batch_size,(imaDim,imaDim))
```

### Model 1

```python
imaDim =128
numClass = 6

def getModelArch(imaDim=128):
    model = Sequential()
    model.add(Lambda(lambda x: x/127.5-0.5, input_shape=(3, imaDim, imaDim)))

    model.add(Conv2D(32, 3, 3))
    model.add(Activation('relu'))
    model.add(BatchNormalization(axis=1))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(64, 3, 3))
    model.add(Activation('relu'))
    model.add(BatchNormalization(axis=1))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(128, 3, 3))
    model.add(Activation('relu'))
    model.add(BatchNormalization(axis=1))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors

    model.add(Dense(256))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    model.add(Dense(64))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    model.add(Dense(numClass))
    model.add(Activation('softmax'))
    return model
```

```python
model = getModelArch(128)
model.summary()
```

```python
ct = time.time()

model = getModelArch(imaDim)

model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

mhist= model.fit_generator(
        generator=train_generator,
        samples_per_epoch=100* batch_size,
        nb_epoch=10,
        validation_data=test_generator,
        nb_val_samples= 100 * batch_size,
        callbacks=[TensorBoard(log_dir='/tmp/autoencoder')])
print("time= {} min".format((time.time()-ct)/60.))
```

```python
plt.figure(figsize=(15,4))
#tr, val = 'loss', 'val_loss'
#tr, val = 'acc', 'val_acc'

metric = ['loss','acc']

for k in range(len(metric)):
    tr, val = metric[k], "val_"+metric[k]
    plt.subplot(1,2,k+1)
    plt.plot(mhist.epoch, mhist.history[tr],label=tr)
    plt.plot(mhist.epoch, mhist.history[val],label=val)
    plt.legend()
```
